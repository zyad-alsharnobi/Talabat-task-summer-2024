```python
import pandas as pd
import numpy as np
```

**1- Reading The data**

```python
df = pd.read_csv(r"Nasr City Restaurants.csv", encoding="utf-8")
df
```

**2-Remove Duplicate Titles**

```python
df.sort_values('reviewsCount', ascending=False,inplace=True)
```

```python
df.drop_duplicates(subset=['title','location/lat','location/lng'],keep='first',inplace=True)
```

```python
df['cuisine'].unique()
```

**3-Flag Koshary restaurants**

```python
import re

pattern = re.compile(r'\bkoshary\b|\bkoshari\b|\bkushari\b|\bkushary\b', re.IGNORECASE)

df['Is_Koshary'] = df['cuisine'].apply(lambda x: bool(pattern.search(str(x))))

df
```

```python
df[df['Is_Koshary'] == True] 
```

**Another way**

```python
koshary_keywords = ['koshary', 'koshari','kushary','kushari','كشري']
df['Koshary'] = df['title'].str.contains('|'.join(koshary_keywords), case=False, na=False)
```

```python
df[df['Koshary'] == True]
```

```python
koshary_df = df[df['Koshary']]
```

```python
min = float('inf')
pair = (None,None)

for i in range(len(koshary_df)):
    for j in range(i+1, len(koshary_df)):
        lat1, lon1 = koshary_df.iloc[i]['location/lat'], koshary_df.iloc[i]['location/lng']
        lat2, lon2 = koshary_df.iloc[j]['location/lat'], koshary_df.iloc[j]['location/lng']
        
        euclidean_distance = np.sqrt((lat1 - lat2)**2 + (lon1 - lon2)**2)
        
        if euclidean_distance < min:
            min = euclidean_distance
            pair = (koshary_df.iloc[i]['title'], koshary_df.iloc[j]['title'])
            
print(f"The two closest Koshary restaurants are {pair[0]} and {pair[1]} with a distance of {min:.4f}")
```
