```python
import pandas as pd
import re
```

```python
df = pd.read_csv(r'New_Cairo_Restaurants.csv')
df.head()
```

```python
def lat_long(data):
    exp = re.search(r'!3d(-?\d+\.\d+)!4d(-?\d+\.\d+)',data)
    if exp:
        return float(exp.group(1)), float(exp.group(2))
    return None, None
```

```python
df[['location/lat','location/lng']] = df['url'].apply(lambda x: pd.Series(lat_long(x)))
```

```python
df
```

```python
df['reviewsCount'] = df['reviewsCount'].str.replace('-','')
df
```

```python
df['title'] = df['title'].str.replace('A-Za-z-9','')
df
```

```python
df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
df.to_csv(r'New_Cairo_Restaurants.csv',index=False)
```
